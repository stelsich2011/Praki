```python
# Импортирование библиотек
import cv2
import numpy as np
```

```python
# Загрузка изображения
img = cv2.imread('inputs/input.png', cv2.IMREAD_GRAYSCALE)
```

```python
# Гаусс
def gauss(img):
    return cv2.GaussianBlur(img, (5, 5), 0)
```

```python
# Собель
def sobel(img, kernelx, kernely):
    img_blur = gauss(img)
    res1 = cv2.filter2D(img_blur, cv2.CV_64F, kernelx)
    res2 = cv2.filter2D(img_blur, cv2.CV_64F, kernely)
    magnitude = np.sqrt(res1**2 + res2**2)
    magnitude = np.uint8(np.absolute(magnitude))
    return magnitude
kernelx = np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]]) # Gx
kernely = np.array([[-1, -2, -1], [0, 0, 0], [1, 2, 1]]) # Gy
```

```python
# Мощность
def power_and_orientation(img, kernelx, kernely):
    img_blur = gauss(img)
    res1 = cv2.filter2D(img_blur, -1, kernelx)
    res2 = cv2.filter2D(img_blur, -1, kernely)
    power = np.sqrt(res1 ** 2 + res2 ** 2) # мощность
    orientation = np.arctan2(res2, res1) # направление
    return power, orientation
```

```python
# NMS
def nms(img):
    return cv2.Canny(img, 70, 180)
```

```python
# Двойнойм порог
def double_porog(img):
    return cv2.Canny(img, 50, 150)
```

```python
# Вызов функций
cv2.imwrite('results/gauss.png', gauss(img)) # Результат гаусса
cv2.imwrite('results/sobel.png', sobel(img, kernelx, kernely)) # Результат собеля
cv2.imwrite('results/nms.png', nms(img)) # Результат nms
cv2.imwrite('results/res.png', double_porog(nms(img))) # Результат двойного порога
```
