```python
# Импортирование библиотек
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
X, y = make_classification(n_samples=150, n_features=2, n_informative=2, n_classes=3, n_redundant=0, n_clusters_per_class=1, random_state=54)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=50, random_state=54)
```

```python
model = LogisticRegression()
model.fit(X, y)
```

```python
h = .02  # шаг сетки
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.figure(figsize=(10, 6))
plt.pcolormesh(xx, yy, Z, cmap=plt.cm.Paired, alpha=0.3)
plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k', cmap=plt.cm.Paired)
plt.show()

plt.show()
```
