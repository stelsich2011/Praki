```python
import cv2
from ultralytics import YOLO
```

```python
model = YOLO('yolo26m.pt')
cap = cv2.VideoCapture(0)
metrics_before = model.val(data='dataset/data.yaml')
print(metrics_before.box.map50)
print(metrics_before.box.map)
```

```python
model.train(data='dataset/data.yaml', epochs=10, imgsz=640, device=0, batch=4)

metrics_after = model.val()
print(metrics_after.box.map50)
print(metrics_after.box.map)
```
